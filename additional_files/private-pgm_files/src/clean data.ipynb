```python
import pandas as pd
```

```python
df = pd.read_csv('data/propublica_data_for_fairml.csv')
df.head()
```

```python
df['Number_of_Priors'].nunique()
```

```python
df['Asian'] = df['Asian'].apply(lambda x: x*2)
df['Hispanic'] = df['Hispanic'].apply(lambda x: x*3)
df['Native_American'] = df['Native_American'].apply(lambda x: x*4)
df['Other'] = df['Other'].apply(lambda x: x*5)
df.head()
```

```python
df['Race'] = df['African_American'] + df['Asian'] + df['Hispanic'] + df['Native_American']  +df['Other']
df.head()
```

```python
df = df.drop(['African_American', 'Asian','Hispanic' ,'Native_American' , 'Other'], axis=1)
df.head()
```

```python
df.to_csv('data/cleaned_Compas.csv')  
```
